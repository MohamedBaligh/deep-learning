```python
from keras.layers import Conv2D,Dense
```

```python
from keras.models import Sequential
```

```python
from skimage.io import imshow,imread
from skimage.transform import resize
```

```python
model = Sequential()
model.add(Conv2D(2,(3,3),input_shape=(720,1280,3)))
model.add(Conv2D(30,(3,3)))
model.summary()
```

```python
imshow('rg.jpg')
```

```python
img = imread('rg.jpg')
```

```python
img.shape
```

```python
out = model.predict(img.reshape(1,720,1280,3))
```

```python
out.shape
```

```python
out = out/255.0
```
